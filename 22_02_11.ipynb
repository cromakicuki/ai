```python
# (영어) 알파벳과 숫자를 섞어서 입력을 받아서 숫자와 문자를
# 분리하여 출력하는 프로그램을 작성하시오.

n = input('숫자를 포함한 문자열을 입력하세요 >> ')
al = ''
no = ''

for i in n:
  if i.isalpha():
    al = al + i
  elif i.isdigit():
    no = no + i
    

print(f'문자: {al}')
print(f'숫자: {no}')
```

```python
# 문자와 숫자의 개수를 각각 포함하여 출력

n = input('숫자를 포함한 문자열을 입력하세요 >> ')
al = ''
no = ''

for i in n:
  if i.isalpha():
    al = al + i
  elif i.isdigit():
    no = no + i
    

print(f'문자({len(al)}): {al}')
print(f'숫자({len(no)}): {no}')
```

```python
# 문자와 숫자의 개수를 각각 포함하여 출력하세요
# 단 isalpha() isdigit() int() 함수를 사용하지 않고 작성하시오
# 

# 숫자를 포함한 문자열을 입력하세요 >>> 1234asdf324as
# 문자(6) :  asdfas
# 숫자(7) :  1234324

# 함수형태가 안 들어가면 감점... (def)
# 행마다 설명(주석) 넣기

n = input('숫자를 포함한 문자열을 입력하세요 >> ')
al = ''
no = ''

for i in n:
  if i >= 'A' and i <= 'Z':
    al = al + i
  elif i >= 'a' and i <= 'z':
    al = al + i
  elif i >= '0' and i <= '9':
    no = no + i
    
print(f'문자({len(al)}): {al}')
print(f'숫자({len(no)}): {no}')
```

```python
#######################################################################################
# 알고리즘
#######################################################################################
# 알고리즘은 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어나 생각
# 알고리즘 표현 방법은 수도코드(pseudo code), 순서도(flow chart)

# 프로그램 작성 순서
# 기획 - 설계(알고리즘) - 프로그래밍 - 디버그


# 좋은 알고리즘
# 알기 쉽고, 속도가 빠르고, 효율적이고(메모리를 적게 사용), 재이용이 쉽다.

# 왜 알고리즘? (을 써야하는가)
# 좋은 프로그램을 만들 수 있고, 좋은 프로그램인지 판단할 수 있고
# 프로그램 작성을 효율적으로 할 수 있고 프로그래밍 기술을 향상시키는데 큰 도움이 된다.

# 좋은 알고리즘의 조건
# 결과를 얻을 수 있어야 하고 반드시 종료되어야 한다.

# 알고리즘 세가지 기본형
# 1 순차구조 : 처음부터 순서대로 처리 (가장 많이 씀)
# 2 선택구조 : 조건식 if
# 3 반복구조 : 반복문 for while


# 순서도 flow 차트, 흐름도
# 프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법
# 처리와 절차를 시각적으로 표현한다.
# 자신이 생각하고 있는 알고리즘이 올바른지, 처리와 절차에 문제는 없는지를 
# 눈으로 확인할 수 있고 다른 사람에게도 쉽게 전달할 수 있는 특징이 있다.
# 주의 사항
# 처리 기호나 판단 기호가 많아지면 흐름이 복잡하게 꼬인다.
# 1. 입력은 반드시 위에서 들어가도록 만든다.
# 2. 판단기호로부터 나오는 흐름선에는 Yes No를 확실하게 기입한다.
# 3. 입력선이나 출력선이 없는 처리 기호는 만들지 않는다.
```
